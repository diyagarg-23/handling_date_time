WORKING WITH DATES

```python
#converting to datetime datatype
date['date'] = pd.to_datetime(date['date'])
```

```python
date.info()
```

1.extract year

```python
date['date_year'] = date['date'].dt.year
date.head()
```

2.extract month

```python
date['date_month'] = date['date'].dt.month

date.head()
```

```python
date['date_month_name'] = date['date'].dt.month_name()
date.head()
```

3.extract day

```python
date['date_day'] = date['date'].dt.day

date.head()
```

```python
#day of week
date['date_dayofweek'] = date['date'].dt.dayofweek
date.head()
```

```python
#day of week - name
date['date_day_name'] = date['date'].dt.day_name()

date.drop(columns=['product_id','city_id','orders']).head()
```

```python
#is it a weekend?
date['date_is_weekend'] = np.where(date['date_day_name'].isin(['Sunday','Saturday']),1,0)

date.drop(columns =['product_id','city_id','orders']).head()
```

Extract week of year

```python
# Correct way to get the week number
date['date_week'] = date['date'].dt.isocalendar().week

# Drop the specified columns and display the first few rows
date.drop(columns=['product_id', 'city_id', 'orders']).head()
```

extract quarter

```python
date['date_quarter'] = date['date'].dt.quarter

date.drop(columns=['product_id','city_id','orders']).head()
```

extract semester

```python
date['semester'] = np.where(date['date_quarter'].isin([1,2]),1,2)

date.drop(columns=['product_id','city_id','orders']).head()
```

extract time elapsed between dates

```python
import datetime
```

```python
today = datetime.datetime.today()#datetime class in datetime module

today
```

```python
today-date['date']
```

```python
(today-date['date']).dt.days
```

```python
#months passed

np.round((today - date['date']).dt.days / 30.44, 0)
```
